## Task 1- Dealing with Null values and Duplicates
- There are 10 null values in the data set. This implies that there are 10 rows with missing data.
- Since the number of missing values is less than 1% of the entire dataset, the approach to deal with
    the missing values is to delete them using the dropna() method


```python
# Check for Duplicates
df.duplicated().sum()
```

```python
# Checking for null values
df.isnull().sum()
```

```python
print(f'Total number of null values = {df.isnull().sum().sum()}')
```

```python
# Deleting Null Values
df = df.dropna(how='any')
df.info()
```
